```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.scatter(x='tip',y='total_bill',data=df,c='g',s=30,label = 'bill')
plt.scatter(x='tip',y='size',data=df,c='r',marker='+',label='size')
plt.legend(title='bill')
plt.xlabel('Tip',fontsize=14)
plt.ylabel('bill and size',fontsize=14)
```

```python
sns.lmplot(x='tip',y='total_bill',data=df,fit_reg=False)
```

```python
plt.figure
df.tip.plot(kind='hist')
```

```python
plt.figure
sns.distplot(df.tip)
```

```python
plt.figure
sns.distplot(df.tip,kde=False)
```

```python
plt.figure
df.sex.value_counts().plot(kind='bar')
```

```python
sns.countplot(df.sex)
```

```python
plt.figure
df.groupby(['time','day'])['tip'].mean().plot(kind='bar')
```

```python
plt.figure
df.groupby(['time','day'])['tip'].agg(['mean','std']).plot(kind='bar')
```

```python
plt.figure
df.groupby(['time','day'])['tip'].agg(['mean','std']).plot(kind='bar',stacked=True)
```

```python
plt.figure
df.groupby(['time','day'])['tip'].agg(['mean','std']).plot(kind='barh',stacked=True)
```

```python
plt.figure
sns.boxplot(df.tip)
```

```python
sns.lmplot(x='tip',y='total_bill',data=df,fit_reg=False,hue='sex',col='time',row='day')
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
sns.boxplot(df.tip, ax=ax2)
sns.boxplot(df.total_bill,ax=ax1)
ax1.set_xlabel('total ')
ax1.set_ylabel('total bill')
ax2.set_ylabel('tip')
```
